```python
import numpy as np
import pandas as pd
np.random.seed(42)

def generate_random_samples(num_samples):
    # Define the range for each SLURM priority parameter
    # Assuming weights can vary between 0 and 100,000 for simplicity
    min_weight, max_weight = 0, 100000

    # Generate random samples for each parameter
    priority_weight_age = np.random.uniform(min_weight, max_weight, num_samples)
    priority_weight_job_size = np.random.uniform(min_weight, max_weight, num_samples)
    priority_weight_fairshare = np.random.uniform(min_weight, max_weight, num_samples)

    # Create a DataFrame to store these values
    data = {
        'PriorityWeightAge': priority_weight_age,
        'PriorityWeightJobSize': priority_weight_job_size,
        'PriorityWeightFairshare': priority_weight_fairshare
    }
    df = pd.DataFrame(data)

    return df

# Generate 100 random samples
num_samples = 100
sample_data = generate_random_samples(num_samples)

# Display the first few rows of the generated data
print(sample_data.head())

# Optionally, save the data to a CSV file
sample_data.to_csv('slurm_priority_parameters_samples.csv', index=False)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the data using the previous function or load your dataset
num_samples = 100
sample_data = generate_random_samples(num_samples)

# Creating the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(sample_data['PriorityWeightAge'], 
           sample_data['PriorityWeightJobSize'], 
           sample_data['PriorityWeightFairshare'], 
           c='blue', marker='o', alpha=0.5)

ax.set_xlabel('Priority Weight Age')
ax.set_ylabel('Priority Weight Job Size')
ax.set_zlabel('Priority Weight Fairshare')

plt.title('3D Scatter Plot of SLURM Parameters')
plt.savefig('3D_Scatter_SLURM_Parameters.png')
plt.show()
```
